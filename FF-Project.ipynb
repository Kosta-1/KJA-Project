```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%pylab inline

df = pd.read_csv("Fast-Food-Data.csv")
```

```python
# Restaurants
#----------------------------------------------------------------------------
def burger_king():
    df_burger_king = df.loc[df["Fast Food Restaurant"] == "Burger King"]
    return df_burger_king

def mcdonalds():
    df_burger_king = df.loc[df["Fast Food Restaurant"] == "McDonald's"]
    return df_burger_king

def mcdonalds():
    df_mcdonalds = df.loc[df["Fast Food Restaurant"] == "McDonald's"]
    return df_mcdonalds

def wendys():
    df_wendys = df.loc[df["Fast Food Restaurant"] == "Wendy's"]
    return df_wendys

def white_castle():
    df_white_castle = df.loc[df["Fast Food Restaurant"] == "White Castle"]
    return df_white_castle

def whataburger():
    df_whataburger = df.loc[df["Fast Food Restaurant"] == "Whataburger"]
    return df_whataburger

def in_n_out_burger():
    df_in_n_out_burger = df.loc[df["Fast Food Restaurant"] == "In-N-Out Burger"]
    return df_in_n_out_burger

def hardees():
    df_hardees = df.loc[df["Fast Food Restaurant"] == "Hardee's"]
    return df_hardees

def jack_in_the_box():
    df_jack_in_the_box = df.loc[df["Fast Food Restaurant"] == "Jack in the Box"]
    return df_jack_in_the_box

def chickfila():
    df_chickfila = df.loc[df["Fast Food Restaurant"] == "Chick-fil-A"]
    return df_chickfila

def dairy_queen():
    df_dairy_queen = df.loc[df["Fast Food Restaurant"] == "Dairy Queen"]
    return df_dairy_queen

def sonic():
    df_sonic = df.loc[df["Fast Food Restaurant"] == "Sonic"]
    return df_sonic

def carls_jr():
    df_carls_jr = df.loc[df["Fast Food Restaurant"] == "Carl's Jr."]
    return df_carls_jr
#----------------------------------------------------------------------------



def low_cal():
    df_low_cal = df.sort_values("Calories",ascending = True)
    lowest_calories = pd.DataFrame(df_low_cal.head(10))
    return lowest_calories
def high_cal():
    df_high_cal = df.sort_values("Calories",ascending = False)
    highest_calories = pd.DataFrame(df_high_cal.head(10))
    return highest_calories
def low_sodium():
    df_low_sodium = df.sort_values("Sodium (mg)",ascending = True)
    lowest_sodium = pd.DataFrame(df_low_sodium.head(10))
    return lowest_sodium
def high_sodium():
    df_high_sodium = df.sort_values("Sodium (mg)",ascending = False)
    highest_sodium = pd.DataFrame(df_high_sodium.head(10))
    return highest_sodium

print("1. Search a specific restaurant: Burger King, McDonalds, Wendy's, White Castle, Whataburger, In-N-Out Burger, Hardee's, Chick-fil-A, Jack in the Box, Sonic, Dairy Queen, Carl's Jr.")
print("2. Search for a specific food: Burger, French Fries, Milkshake, Chicken Nuggets, Grilled Chicken Sandwich, Breaded Chicken Sandwich ")
print("3. Search for Top 10 food items with lowest calories: Type 'Low Calories'")
print("4. Search for Top 10 food items with highest calories: Type 'High Calories'")
print("5. Search for Top 10 food items with lowest sodium: Type 'Low Sodium'")
print("6. Search for Top 10 food items with highest sodium: Type 'High Sodium'")

user_input = input("Enter what you want to search: ")
user_input = user_input.lower()
import string
user_input = user_input.translate(user_input.maketrans('', '', string.punctuation))

while True:
#Low_Calories
    if user_input == "low calories":
        final_low_cal = low_cal()
        display(final_low_cal)
        break
#High Calories
    elif user_input == "high calories":
        final_high_cal = high_cal()
        display(final_high_cal)
        break
#Low Sodium    
    elif user_input == "low sodium":
        final_low_sodium = low_sodium()
        display(final_low_sodium)
        break
#High Sodium        
    elif user_input == "high sodium":
        final_high_sodium = high_sodium()
        display(final_high_sodium)
        break
# Fast Food Restaurants :
    elif user_input == "burger king":
        final_burger_king = burger_king()
        display(final_burger_king)
        break    
    elif user_input == "mcdonalds":
        final_mcdonalds = mcdonalds()
        display(final_mcdonalds)
        break
        
    elif user_input == "wendys":
        final_wendys = wendys()
        display(final_wendys)
        break
        
    elif user_input == "white castle":
        final_white_castle = white_castle()
        display(final_white_castle)
        break
        
    elif user_input == "whataburger":
        final_whataburger = whataburger()
        display(final_whataburger)
        break
        
    elif user_input == "innoutburger":
        final_innoutburger = in_n_out_burger()
        display(final_innoutburger)
        break
        
    elif user_input == "hardees":
        final_hardees = hardees()
        display(final_hardees)
        break
        
    elif user_input == "jack in the box":
        final_jack_in_the_box = jack_in_the_box()
        display(final_jack_in_the_box)
        break
        
    elif user_input == "chickfila":
        final_chickfila = chickfila()
        display(final_chickfila)
        break
        
    elif user_input == "dairy queen":
        final_dairy_queen = dairy_queen()
        display(final_dairy_queen)
        break
        
    elif user_input == "sonic":
        final_sonic = sonic()
        display(final_sonic)
        break
        
    elif user_input == "carls jr":
        final_carls_jr = carls_jr()
        display(final_carls_jr)
        break
  

    else:
        print("1. Search a specific restaurant: Burger King, McDonalds, Wendy's, White Castle, Whataburger, In-N-Out Burger, Hardee's, Chick-fil-A, Jack in the Box, Sonic, Dairy Queen, Carl's Jr.")
        print("2. Search for a specific food: Burger, French Fries, Milkshake, Chicken Nuggets, Grilled Chicken Sandwich, Breaded Chicken Sandwich ")
        print("3. Search for Top 10 food items with lowest calories: Type 'Low Calories'")
        print("4. Search for Top 10 food items with highest calories: Type 'High Calories'")
        print("5. Search for Top 10 food items with lowest sodium: Type 'Low Sodium'")
        print("6. Search for Top 10 food items with highest sodium: Type 'High Sodium'")
        user_input = input("Enter what you want to search: ")
```
